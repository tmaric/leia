```python
from itertools import zip_longest
template = '3Ddeformation'
case = ['E_GEOM_ALPHA', 'O(E_GEOM_ALPHA)', 'O_LOCAL(E_GEOM_ALPHA)']

def database_smallest(df, template, case):
    endTime = df.loc[(df[('database','TEMPLATE')]==template), ('case','TIME')].max()
    view_df = df[(df[('database','TEMPLATE')]==template) & (df[('case','TIME')]==endTime)]

    mi = df.columns
    database = mi[mi.get_locs(['database'])].to_list()
    studyparameters = mi[mi.get_locs(['studyparameters'])].to_list()
    case = list(zip_longest([], case, fillvalue='case'))

    viewlargest_df = view_df.nsmallest(10, case[0])
    return viewlargest_df[database + studyparameters + case ]

database_smallest(df, template, case)
```

```python
template = '3Drotation'
case = ['E_GEOM_ALPHA', 'O(E_GEOM_ALPHA)', 'O_LOCAL(E_GEOM_ALPHA)']

database_smallest(df, template, case)
```
